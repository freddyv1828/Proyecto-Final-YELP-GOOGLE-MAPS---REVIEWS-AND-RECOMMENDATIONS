```python
# @title Example 1: Query a table with SQL and magic commands
# Running this code will query a table in BigQuery and download
# the results to a Pandas DataFrame named `results`.
# Learn more here: https://cloud.google.com/bigquery/docs/visualize-jupyter

%%bigquery results --project eminent-cycle-415715
SELECT * FROM `eminent-cycle-415715.prueba_ai_ML.business` #this table name was set based on the table you chose to query
```

```python
# @title Example 2: Query a table with BigQuery DataFrames
# With BigQuery DataFrames, you can use many familiar Pandas methods, but the
# processing happens BigQuery rather than the runtime, allowing you to work with larger
# DataFrames that would otherwise not fit in the runtime memory.
# Learn more here: https://cloud.google.com/python/docs/reference/bigframes/latest

import bigframes.pandas as bf

bf.options.bigquery.location = "us-central1" #this variable is set based on the dataset you chose to query
bf.options.bigquery.project = "eminent-cycle-415715" #this variable is set based on the dataset you chose to query
```

```python
df = bf.read_gbq("eminent-cycle-415715.prueba_ai_ML.business") #this variable is set based on the dataset you chose to query
```

```python
# BigFrames can work with tables that are too large to fit in the notebook memory.
# Look at the first 20 rows.
df.head(20)
```

```python
# Bloque 4: Visualización de distribuciones y relaciones entre variables
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Convertir el DataFrame de BigQuery a un DataFrame de Pandas
df_pd = df.to_pandas()

# Ejemplo de visualización de la distribución de una variable numérica
plt.figure(figsize=(10, 6))
sns.histplot(data=df_pd, x='stars', bins=20)
plt.title('Distribución de stars')
plt.xlabel('Stars')
plt.ylabel('Frecuencia')
plt.show()

# Ejemplo de visualización de una relación entre dos variables
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_pd, x='stars', y='review_count')
plt.title('Relación entre stars y review_count')
plt.xlabel('Stars')
plt.ylabel('Review Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Convertir el DataFrame de BigFrames a un DataFrame de Pandas
df_pd = df.to_pandas()

# Visualizar la distribución de las calificaciones (stars)
plt.figure(figsize=(10, 6))
sns.histplot(data=df_pd, x='stars', bins=20)
plt.title('Distribución de stars')
plt.xlabel('Stars')
plt.show()
```

```python
# Convertir el DataFrame de BigFrames a un DataFrame de Pandas
df_pandas = df.to_pandas()

# Por ejemplo, podemos trazar la distribución de las calificaciones (stars) por categoría de negocio
plt.figure(figsize=(12, 8))
sns.boxplot(data=df_pandas, x='category', y='stars')
plt.title('Distribución de calificaciones por categoría de negocio')
plt.xlabel('Categoría de negocio')
plt.ylabel('Stars')
plt.xticks(rotation=45, ha='right')
plt.show()
```
