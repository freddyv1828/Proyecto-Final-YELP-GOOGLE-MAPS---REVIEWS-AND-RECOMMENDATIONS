```python
import pandas as pd
import funciones
import numpy as np
from datetime import datetime
```

```python
ruta = '../DATASETS/checkin_clean.csv'
ruta_businness = '../DATASETS/business_corregido.csv'
```

```python
# open business
df_business = pd.read_csv(ruta_businness)
```

```python
# Open file
df_checkin = pd.read_csv(ruta)
df_checkin
```

```python
def parse_date(date_str):
    return datetime.strptime(date_str.strip(), '%Y-%m-%d %H:%M:%S')
```

```python
df_checkin['date'] = df_checkin['date'].str.split(',')
df_checkin = df_checkin.explode('date', ignore_index=True)
df_checkin['date'] = df_checkin['date'].apply(parse_date)
df_checkin
```

Obtenemos la lista de ids de los estbalecimientos

```python
df_business_list = df_business["business_id"].tolist()
```

Filtramos los checkins de los establecimientos que nos interesan

```python
df_checkin_filtered = df_checkin[df_checkin["business_id"].isin(df_business_list)]
```
