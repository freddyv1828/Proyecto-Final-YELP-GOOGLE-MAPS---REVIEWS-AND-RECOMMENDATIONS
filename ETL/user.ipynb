```python
import pandas as pd
import funciones
import numpy as np
```

```python
ruta_user = '../DATASETS/user_clean.csv'
```

```python
ruta_review = '../DATASETS/review_filtered.csv'
```

```python
# open
df_user = pd.read_csv(ruta_user)
df_user.head()
```

```python
df_reviews = pd.read_csv(ruta_review)
df_reviews.head()
```

```python
user_id_reviews = df_reviews["user_id"].tolist()
```

Filtramos dataset para obtener solo los usuarios del segemento seleccionado

```python
df_user_filtered = df_user[df_user["user_id"].isin(user_id_reviews)]
```

```python
df_user_filtered = df_user_filtered.drop(columns=["elite", "friends"])
```

```python
df_user_filtered.columns
```

Marcamos como sin nombre los registros nulos de los nombres de los usuarios ya que solo tienen como nulo los nombres y el resto de la nformacion es valiosa

```python
# Suponiendo que quieres reemplazar los nulos con "Sin nombre"
df_user_filtered['name'].fillna('Sin nombre', inplace=True)
```
